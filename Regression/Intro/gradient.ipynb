```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def gradentDescent(x, y) :
    m_curr = b_curr = 0
    learning_rate = 0.08
    iterations = 1000
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        md = -(2/len(x)) * sum(x*(y - y_predicted))
        bd = -(2/len(x)) * sum((y - y_predicted))
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        cost = (1/len(x)) * sum([val** 2 for val in (y- y_predicted)])
        print("m {} , b {}, iteration {}, cost {}".format(m_curr, b_curr, i, cost))
        plt.scatter(m_curr, b_curr)
        plt.plot(m_curr, b_curr , color= 'b')

```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradentDescent(x, y)
```
