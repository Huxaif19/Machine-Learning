```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprice.csv')
df
```

```python
%matplotlib inline
plt.plot(df.area, df.price , color = "black")
plt.xlabel("area(sqrt ft)")
plt.ylabel("price($)")
plt.scatter(df.area, df.price, color = "red" , marker="o")
```

create an linear regression object

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv("areas.csv")
d.head(3)
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```

```python
d
```

```python
d.to_csv("prediction.csv", index = False)
```

```python
import pickle 
```

```python
with open ('model_pickle' , 'wb') as f: 
    pickle.dump(reg,f)
```

```python
with open ('model_pickle', 'rb') as f:
    mp = pickle.load(f)
```
