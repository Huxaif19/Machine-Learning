```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('car.csv')
df.head(5)
```

```python
plt.scatter(df['Mileage'], df['Sell Price($)'], marker='P', color = "r")
plt.plot(df['Mileage'], df['Sell Price($)'], color = "g")
```

```python
x = df[["Mileage", "Age(yrs)"]]
y = df[['Sell Price($)']]
```

```python
X_trian , X_test , Y_train, Y_test = train_test_split(x, y , test_size=0.2, train_size=0.8)
```

```python
len(X_trian)
```

```python
len(X_test)
```

```python
Model = LinearRegression()
```

```python
Model.fit(X_trian, Y_train)
```

```python
Model.predict(X_test)
```

```python
Y_test
```

```python
Model.score(X_test, Y_test)
```
